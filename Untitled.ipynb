```python
#Dependencies
import pandas as pd
```

```python
#Reading in Citi Bike Data CSVs
years=['202101','202102','202103','202104','202001','202002','202003','202004','202005','202006','202007','202008',
      '202009','202010','202011','202012','201901','201902','201903','201904','201905','201906','201907','201908',
      '201909','201910','201911','201912']

citi_bike_data= pd.DataFrame()

for year in years:
    year_df = pd.read_csv(f"JC-{year}-citibike-tripdata.csv")
    
    citi_bike_data = citi_bike_data.append(year_df)
   
    
    print(f"JC-{year}-citibike-tripdata.csv")

citi_bike_data
```

```python
citi_bike_data.to_csv("citi_bike_data.csv")
```
